```python
# Group by state and count positions
state_counts = df.groupby('State')['Position'].count()

# Plot the count of positions by state
state_counts.plot(kind='bar', figsize=(10, 6))
plt.xlabel('State')
plt.ylabel('Count of Positions')
plt.title('Count of Positions by State')
plt.xticks(rotation=90)
plt.show()
```
